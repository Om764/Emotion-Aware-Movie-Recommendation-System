```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import random
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout,Flatten,BatchNormalization,MaxPooling2D,Input
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.layers import Dense,Conv2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data="facial expression dataset/train"
test_data="facial expression dataset/validation"
```

```python
datagen_train = ImageDataGenerator(rescale=1./255)
datagen_test = ImageDataGenerator(rescale=1./255)
```

```python
train_ds = datagen_train.flow_from_directory(
    "facial expression dataset/train",
    target_size=(48,48),
    color_mode="grayscale",
    batch_size=128,
    class_mode="categorical",
    shuffle=True
)

test_ds = datagen_test.flow_from_directory(
    "facial expression dataset/validation",
    target_size=(48,48),
    color_mode="grayscale",
    batch_size=128,
    class_mode="categorical",
    shuffle=True
)

    
```

```python
model1 = Sequential([
    Conv2D(64,(3,3),padding='same',activation="relu", input_shape=(48,48,1)),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),

    Conv2D(128,(5,5),padding='same',activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),

    Conv2D(512,(3,3),padding='same',activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),

    Conv2D(512,(3,3),padding='same',activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),

    Flatten(),

    Dense(256,activation='relu'),
    BatchNormalization(),
    Dropout(0.25),

    Dense(512,activation='relu'),
    BatchNormalization(),
    Dropout(0.25),
    

    Dense(7,activation='softmax')
])


opt=Adam(learning_rate=0.0001)
model1.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
model1.summary()
    

    
           
           
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
checkpoint = ModelCheckpoint("./model1.h5",monitor='val_accuracy', verbose=1,save_best_only=True)

early_stopping=EarlyStopping(monitor='val_loss',min_delta=0,patience=3,verbose=1,restore_best_weights=True)

reduce_learningrate=ReduceLROnPlateau(monitor='val_loss',factor=0.2,patience=3,verbose=1,min_delta=0.0001)

callbacks=[early_stopping,checkpoint,reduce_learningrate]
```

```python
history=model1.fit(train_ds,validation_data=test_ds,epochs=40)
```

```python
data = data[['genres','id','keywords','overview','title','tagline','cast','crew']]
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data.head()
```

```python
data['keywords']=data['keywords'].apply(convert)
```

```python
data['overview']=data['overview'].fillna("")
data['overview']=data['overview'].apply(lambda x: x.split())
```

```python
def cast_convert(obj):
    L=[]
    try:
        counter = 0
        for i in ast.literal_eval(obj):
             if counter != 3:
                 L.append(i["name"])
                 counter+=1
             else:
                 break
    except:
        pass
    return L
        
    
```

```python
data['cast'] = data['cast'].apply(cast_convert)
data.head()
```

```python
data['cast'].iloc[0]
```

```python
data['crew'].iloc[0]
```

```python
def crew_convert(obj):
    L=[]
    try:
        for i in ast.literal_eval(obj):
            if (i['job']) == 'Director' or (i['job']) == 'Producer':
                L.append(i['name'])
    except:
        pass
    return L


                     
```

```python
data['crew']=data['crew'].apply(crew_convert)
```

```python
data.head()
```

```python
data['genres'] = data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords'] = data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data['overview'] = data['overview'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast'] = data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew'] = data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    new_token=[]
    for i in text:
        if i not in stop_words:
            new_token.append(i)
    return new_token
            
```

```python
data['overview'] = data['overview'].apply(preprocess)
```

```python
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['crew'] + data['cast']
```

```python
new_data = data[['id','title','tags']]
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x:" ".join(x))
```

```python
def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemma = [lemmatizer.lemmatize(t) for t in tokens]
    return "".join(lemma)
```

```python
new_data['tags'] = new_data['tags'].apply(lemmatize_text)
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x:x.lower())
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words="english")
from sklearn.cluster import KMeans
```

```python
tfidf_matrix = tfidf.fit_transform(new_data['tags'])
```

```python
kmeans=KMeans(n_clusters=7,random_state=42)
kmeans.fit(tfidf_matrix)
```

```python
new_data.loc[:, 'cluster'] = kmeans.labels_
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import pandas as pd

# Load model
model = load_model("facial_expression.h5")

# Load movie data
movies_df = pd.read_csv("movies_with_clusters.csv")

# Emotion labels (Your training order)
emotion_labels = ['angry','disgust','fear','happy','neutral','sad','surprise']

# Emotion → cluster mapping
emotion_cluster_map = {
    'angry':    [4, 5],
    'disgust':  [1, 3, 6],
    'fear':     [1, 6],
    'happy':    [0, 2, 4],
    'neutral':  [0, 1, 2, 3, 4],
    'sad':      [1, 3, 6],
    'surprise': [0, 2, 4]
}

# Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Preprocess function
def preprocess(face):
    face = cv2.resize(face, (48, 48))
    face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
    face = face.astype("float32") / 255.0
    face = np.expand_dims(face, axis=0)
    return face

# Predict emotion
def get_emotion(face):
    processed = preprocess(face)
    preds = model.predict(processed)[0]
    idx = np.argmax(preds)
    return emotion_labels[idx]

# Recommend movies
def recommend_movies(emotion):
    target_clusters = emotion_cluster_map[emotion]
    movies = movies_df[movies_df["cluster"].isin(target_clusters)]["title"].tolist()
    return np.random.choice(movies, size=min(10, len(movies)), replace=False).tolist()
    #return movies[:10]


# -------- WEBCAM LOOP --------
cap = cv2.VideoCapture(0)

print("Press 'q' to capture emotion...")

captured_face = None

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y), (x+w,y+h), (255,0,0), 2)
        captured_face = frame[y:y+h, x:x+w]

    cv2.imshow("Emotion Detector", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()


# ------ EMOTION + RECOMMENDATION ------
if captured_face is not None:
    emotion = get_emotion(captured_face)
    print("\nDetected Emotion:", emotion)

    recommended = recommend_movies(emotion)
    print("\nRecommended Movies:")
    for m in recommended:
        print("-", m)

else:
    print("No face detected!")
```
